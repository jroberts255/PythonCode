```python
import pandas as pd
import numpy as np
```

```python
Moeller = pd.read_excel('Moeller.xlsx')
```

```python
Moeller['PER'] = round((Moeller['PTS'] + Moeller['AST'] + Moeller['STL'] + Moeller['BLK'] - (Moeller['FGA'] - Moeller['FGM']) - (Moeller['FTA']-Moeller['FTM']) - Moeller['TO'])/(Moeller['GMS']),2)                                                                                      
```

```python
Moeller[['Player', 'PER']].sort_values(by='PER', ascending=False)
```
