```python
import numpy as np
import pandas as pd
```

```python
WinLoss = pd.read_excel('WLData.xlsx')
```

```python
SummarizedWL = WinLoss[['OUTCOME', '1ST Q', '2ND Q', '3RD Q', '4TH Q']]
```

```python
SummarizedWL = SummarizedWL.replace('W', 1)
```

```python
SummarizedWL = SummarizedWL.replace('L', 0)
```

```python
SummarizedWL = SummarizedWL.replace('T', -1)
```

```python
SummarizedWL
```

```python
SummarizedWL['1st Q Pred'] = SummarizedWL['OUTCOME'] - SummarizedWL['1ST Q']
SummarizedWL['2nd Q Pred'] = SummarizedWL['OUTCOME'] - SummarizedWL['2ND Q']
SummarizedWL['3rd Q Pred'] = SummarizedWL['OUTCOME'] - SummarizedWL['3RD Q']
SummarizedWL['4th Q Pred'] = SummarizedWL['OUTCOME'] - SummarizedWL['4TH Q']
```

```python
PredSummarizedWL = SummarizedWL[['1st Q Pred', '2nd Q Pred', '3rd Q Pred', '4th Q Pred']]
```
