```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
moviesdata = pd.read_excel('vgsales.xlsx')
moviesdata
```

```python
topxmovies = moviesdata.loc[((moviesdata['Rank'] >=1)&(moviesdata['Rank']<=5))]
topxmovies
```

```python
fig_dims = (15, 10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x = 'Name', y = 'Value', data = topxmovies, hue = 'SalesType', ax=ax)
```
